#Conclusion
In this notebook, we explored the overall structure of the Walmart demand forecasting dataset and examined the three data sources used throughout the project.

The analysis confirmed that:

- The train dataset contains the target variable (Weekly_Sales) at the Store–Department–Week level.
- The features dataset provides external contextual information such as temperature, fuel price, CPI, unemployment, markdowns, and holiday indicators.
- The stores dataset contains static store attributes, including store type and size.

After understanding the granularity of each dataset, we determined that train should serve as the primary dataset, with features merged using Store and Date, followed by stores using Store.
